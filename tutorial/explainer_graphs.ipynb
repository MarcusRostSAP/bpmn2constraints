# Counterfactual and minimal Necessary 

```python
import sys
import matplotlib.pyplot as plt
sys.path.append('../')
from explainer.explainer_util import Trace, EventLog
from explainer.explainer_regex import ExplainerRegex
import time


# Initialize the explainer
explainer = ExplainerRegex()
explainer.add_constraint('^B')

# Define trace lengths
trace_lengths = [10, 100, 200, 500, 1000]
traces = [Trace(['A', 'B', 'C'] * length) for length in trace_lengths]  # Simple pattern repeated to match length

minimal_explanations_times = []
counterfactual_explanations_times = []

# Measure and store times
for trace in traces:
    start_time = time.time()
    explainer.minimal_expl(trace)
    minimal_explanations_times.append(time.time() - start_time)

    start_time = time.time()
    explainer.counterfactual_expl(trace)
    counterfactual_explanations_times.append(time.time() - start_time)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(trace_lengths, minimal_explanations_times, 'bo-', label='Minimal Explanation Time')
plt.plot(trace_lengths, counterfactual_explanations_times, 'ro-', label='Counterfactual Explanation Time')
plt.xlabel('Trace Length')
plt.ylabel('Time (seconds)')
plt.title('Trace Length vs. Explanation Calculation Time')
plt.legend()
plt.grid(True)
plt.show()
```
